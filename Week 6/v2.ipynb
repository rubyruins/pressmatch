```python
beats = []
```

```python
def find_beats(url):
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    sauce = urlopen(req).read()
    soup = bs.BeautifulSoup(sauce,'lxml')
    try:
        beat = []
        temp = soup.find(class_ = 'person-details-beats').find_all('a')
        for i in temp:
            beat.append(i.getText())
        beat = ', '.join(beat)
    except:
        beat = '-'
    return beat
```

```python
for i in range(len(data)):
    try:
        beats.append(find_beats(data.iloc[i].redirect_url))
    except:
        beats.append('-')
    if i % 100 == 0:
        print(i)
```
