```python
outlets = outlets.dropna().reset_index()
```

```python
outlets = outlets[['Name', 'Muckrack', 'Webpage']]
```

```python
outlets
```

```python
for i in range(len(outlets)):
    url = outlets.iloc[i].Muckrack
    try:
        req = Request(url, headers = {'User-Agent': 'Mozilla/5.0'})
        sauce = urlopen(req).read()
        soup = bs.BeautifulSoup(sauce,'lxml')
        link = soup.find_all(class_ = 'mr-contact-item-inner')[0].text
    except:
        link = "Cannot find"
    outlets.iloc[i].Webpage = link
```

### Get popularity

```python
outlets = outlets[outlets.Webpage != "Cannot find"].reset_index()
```

```python
outlets['Popularity'] = ['0'] * 132
outlets = outlets[["Name", "Muckrack", "Webpage", "Popularity"]]
```

```python
outlets
```

```python
for i in range(len(outlets)):
    url = outlets.iloc[i].Webpage
    try:
        xml = urlopen('http://data.alexa.com/data?cli=10&dat=s&url={}'.format(url)).read() 
        text = int(re.search(r'<POPULARITY[^>]*TEXT="(\d+)"', str(xml)).groups()[0])
    except:
        text = "Cannot find"
    outlets.iloc[i].Popularity = text
```

```python
outlets = outlets[outlets.Popularity != "Cannot find"].reset_index()
outlets = outlets[["Name", "Muckrack", "Webpage", "Popularity"]]
```

```python
outlets.Popularity = outlets.Popularity.astype('int')
```

```python
outlets = outlets.sort_values('Popularity').reset_index()[["Name", "Muckrack", "Webpage", "Popularity"]]
```
