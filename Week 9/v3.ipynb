```python
mapped_tiers = []
mapped_correctly = []

for i in range(len(results)):
    beats = ast.literal_eval(results.iloc[i].beats)
    # print(beats)
    temp = []
    for beat in beats:
        j = mapping[beat.strip()]
        temp = temp + j
    # print(temp)
    print(beats, '\n', temp, '\n', results.iloc[i].predicted)
    correct = True if set(ast.literal_eval(results.iloc[i].predicted)).intersection(set(temp)) else False
    print(correct)
    mapped_tiers.append(temp)
    mapped_correctly.append(correct)
    print()
```

```python
results['mapped_tiers'] = mapped_tiers
results['mapped_correctly'] = mapped_correctly
```
