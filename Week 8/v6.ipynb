```python
mapping = {'Arts and Entertainment':['Events and Attractions', 'Fine Art', 'Pop Culture', 'Books and Literature', 'Movies', 'Music and Audio'],
'Beauty':['Style & Fashion', 'Shopping'],
'Business and Finance':['Business and Finance', 'Personal Finance'],
'Crime and Justice':['News and Politics'],
'Education':['Education', 'Careers'],
'Energy':['Automotive', 'Technology & Computing'],
'Environment':['Home & Garden', 'Healthy Living'],
'Fashion':['Style & Fashion', 'Shopping'],
'Food and Dining':['Food & Drink'],
'Health':['Healthy Living', 'Medical Health'],
'Media':['Movies', 'Music and Audio', 'Pop Culture', 'Video Gaming', 'Television'],
'Military':['News and Politics'],
'Opinion and Editorial':['News and Politics'],
'Politics':['News and Politics'],
'Real Estate':['Real Estate'],
'Religion':['Religion & Spirituality'],
'Science':['Science'],
'Sports':['Sports'],
'Technology':['Technology & Computing'],
'Transportation':['Travel'],
'Travel':['Travel', 'Events and Attractions'],
'Weather':['Science', 'Home & Garden'],
'World':['News and Politics', 'Technology & Computing']}
```

```python
mapped_tiers = []
mapped_correctly = []
```

```python
for i in range(len(results)):
    beats = ast.literal_eval(results.iloc[i].beats)
    # print(beats)
    temp = []
    for beat in beats:
        j = mapping[beat.strip()]
        temp = temp + j
    # print(beats, '\n', temp, '\n', results.iloc[i].predicted)
    correct = results.iloc[i].predicted in temp
    # print(correct)
    mapped_tiers.append(temp)
    mapped_correctly.append(correct)
    # print()
```

```python
results['mapped_tiers'] = mapped_tiers
results['mapped_correctly'] = mapped_correctly
```
