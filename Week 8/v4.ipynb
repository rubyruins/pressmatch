```python
mapped_tier1 = []

for i in range(len(results)):
    beats = results.iloc[i].beat.split(',')
    print(i, beats)
    mapped_tier_list = [mapping[beat.strip()] for beat in beats]
    print(mapped_tier_list)
    mapped_tier1.append(mapped_tier_list)
```

```python
mapped_correctly = []

for i in range(len(results)):
    actual_tier1 = results.iloc[i].actual_tier1
    mapped_tiers = results.iloc[i].mapped_tier
    if actual_tier1 == 'None':
        mapped_correctly.append(False)
    else:
        if actual_tier1 in mapped_tiers:
            mapped_correctly.append(True)
        else:
            mapped_correctly.append(False)
```

```python
results['mapped_correctly'] = mapped_correctly
```
